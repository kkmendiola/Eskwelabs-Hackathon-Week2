```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('applications.csv')
df
```

```python
df.describe()
```

```python
df_loan = df.dropna(subset=['LOAN_DEFAULT'])
```

```python
df_loan.describe() 
```

```python
df_clean = df_loan.drop(['Unnamed: 0', 'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21', 'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR'], axis=1)
df_clean
```

```python
df_loan.drop(df_loan.iloc[:, 32:223], inplace = True, axis = 1) 
```

```python
df_loan['AGEYEAR']=df_loan['DAYS_BIRTH']/-365
df_loan.AGEYEAR = df_loan.AGEYEAR.astype(int)
```

```python
#Distribution Plot of Age Groups (Grouped in Increments of 5)

AgeGrid = sns.FacetGrid(df_loan, col='LOAN_DEFAULT', hue="LOAN_DEFAULT", palette=["#779ecb", "#ff6961"], height=4)
AgeGrid.map(sns.distplot, "AGEYEAR",  bins = np.arange(21, 69, 5)).set_axis_labels('Age (in increments of 5)')
#plt.title('Frequency Distribution by Age Segments (increments of 5) for Good and Bad Loans')


#plt.xlabel('Age (grouped in increments of 5)')
#plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
#plt.yticks(np.arange(0, 81, 10))
#plt.legend((p1[0], p2[0]), ('Men', 'Women'))
```

```python
#Frequency of Loans by Quality and Car Ownership Status
tab = pd.crosstab(df_loan.OWN_CAR, df_loan.LOAN_DEFAULT)
tab.plot(kind="bar", figsize=(8,8),stacked=True,  color=["#779ecb", "#ff6961"])
plt.title('Frequency of Loans by Quality and Car Ownership Status', fontsize=20)
plt.xlabel('Ownership of Car Status')
plt.ylabel('Count (in number of loans)')

#1331/13998: 9.5% of Non-car owners that loan money succumb to default.
#615/7453: 8.3% of Car owners that loan money succumb to default
```

```python
#Frequency of Loans by Quality and Occupation
tab_occ = pd.crosstab(df_loan.OCCUPATION_TYPE, df_loan.LOAN_DEFAULT)
tab_occ.plot(kind="bar", figsize=(8,8),stacked=True, color=["#779ecb", "#ff6961"])
plt.xlabel('Occupation')
plt.ylabel('Count (in number of loans)')
plt.title('Frequency of Loans by Quality and Occupation')
```

```python
#Frequency of Good Loans by Occupation
LOAN=df_loan.loc[df_loan.LOAN_DEFAULT==0,:]
plt.figure(figsize=(12,6))
LOAN.groupby('OCCUPATION_TYPE')['OCCUPATION_TYPE'].count().plot.bar(color="#779ecb")
plt.title('Frequency of Good Loans by Occupation', fontsize=20)
plt.xlabel('Occupation')
plt.ylabel('Count (in number of loans)')
```

```python
#Frequency of Bad Loans by Occupation
DEFAULT=df_loan.loc[df_loan.LOAN_DEFAULT==1,:]
plt.figure(figsize=(12,6))
DEFAULT.groupby('OCCUPATION_TYPE')['OCCUPATION_TYPE'].count().plot.bar(color="#ff6961")
plt.title('Frequency of Bad Loans by Occupation', fontsize=20)
plt.xlabel('Occupation')
plt.ylabel('Count (in number of loans)')
```
