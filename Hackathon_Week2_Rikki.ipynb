```python
LOAN=df_loan.loc[df_loan.LOAN_DEFAULT==0,:]
DEFAULT=df_loan.loc[df_loan.LOAN_DEFAULT==1,:]
plt.figure(figsize=(12,6))
LOAN.groupby('OCCUPATION_TYPE')['OCCUPATION_TYPE'].count().plot.bar()
plt.title('Frequencies of Good Loans by Occupation', fontsize=20)
plt.xlabel('Occupation')
plt.ylabel('Count (in number of loans)')
```

```python
plt.figure(figsize=(12,6))
DEFAULT.groupby('OCCUPATION_TYPE')['OCCUPATION_TYPE'].count().plot.bar()
plt.title('Frequencies of Bad Loans by Occupation', fontsize=20)
plt.xlabel('Occupation')
plt.ylabel('Count (in number of loans)')
```
